```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import sklearn as sk
from sklearn.linear_model import LinearRegression
import statistics
```

```python
crbr = pd.read_csv("Cancer Rates Race.csv")
        
for idx, row in crbr.iterrows():
    if "Unreliable" in row["Mortality-Incidence Age-Adjusted Rate Ratio"]:
        crbr = crbr.drop(idx)
crbr = crbr.dropna()

crbr["Mortality-Incidence Age-Adjusted Rate Ratio"] = crbr["Mortality-Incidence Age-Adjusted Rate Ratio"].astype('float')
crbr["Year"] = crbr["Year"].astype('int')
```

```python
crbr.head()
```

```python
asian = crbr[crbr["Race"] == "Asian or Pacific Islander"]
white = crbr[crbr["Race"] == "White"]
black = crbr[crbr["Race"] == "Black or African American"]
native = crbr[crbr["Race"] == "American Indian or Alaska Native"]
```

```python
years = range(1999,2017)
asian_avgs = []
for year in years:
    vals = asian[asian["Year"]==year]
    avg = statistics.mean(vals["Incidence Age-Adjusted Rate (per 100,000)"].tolist())
    asian_avgs.append(avg)

white_avgs = []
for year in years:
    vals = white[white["Year"]==year]
    avg = statistics.mean(vals["Incidence Age-Adjusted Rate (per 100,000)"].tolist())
    white_avgs.append(avg)
black_avgs = []
for year in years:
    vals = black[black["Year"]==year]
    avg = statistics.mean(vals["Incidence Age-Adjusted Rate (per 100,000)"].tolist())
    black_avgs.append(avg)
native_avgs = []
for year in years:
    vals = native[native["Year"]==year]
    avg = statistics.mean(vals["Incidence Age-Adjusted Rate (per 100,000)"].tolist())
    native_avgs.append(avg)
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.plot(years,asian_avgs, color = "b", label = "Asian or Pacific Islander")
ax.plot(years,white_avgs, color = "r", label = "White")
ax.plot(years,black_avgs, color = "pink", label = "Black or African American")
ax.plot(years,native_avgs, color = "g", label = "American Indian or Alaska Native")
ax.set_xlabel("Year")
ax.set_ylabel("Incidence Age-Adjusted Rate (per 100,000)")
ax.set_xticks(np.arange(1999,2016,2))
ax.legend()
```
